```python
import json
import matplotlib.pyplot as plt


json_file_path = '/content/mydrive/MyDrive/DocTR Dataset/reclistsimage_recall_list_finetuned_db_resnet50_pnew_data.json'
with open(json_file_path, 'r') as file:
    data = json.load(file)
recall_scores = []
for i in range(len(data)):
  recall_scores.append(data[i][0])

plt.figure(figsize=(12, 6))  # Increase the figure size
plt.hist(recall_scores, bins=100, edgecolor='black')
plt.xlabel('Recall Score')
plt.ylabel('Frequency')
plt.title('Recall Score Distribution')
plt.show()
```
